```python
ratings = []

for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    ratings.append(i.text)
    
ratings
```

```python
df = pd.DataFrame({'Titles':titles,'Year_Of_Release':year_of_release, 'Ratings':ratings, })
df
```
