```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
x = data.drop('quality', axis=1)
y = data['quality']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=350)
sc = StandardScaler()
X_scaled= sc.fit_transform(x)
```

```python
data['quality']=data['quality'].replace({7:1,3:0})
data['quality']=data['quality'].replace({8:1,4:0})
data['quality']=data['quality'].replace({5:0})
data['quality']=data['quality'].replace({6:0})
```

```python
for i in data.columns:
    print(data[i].value_counts())
    print("\n")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfr=RandomForestClassifier()
```

```python
rfr.fit(X_train,y_train)
```

```python
pred=rfr.predict(X_test)
pred
```

```python
acc=accuracy_score(y_test,pred)
acc
```
